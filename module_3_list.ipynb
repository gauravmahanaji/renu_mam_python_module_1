```python
list.index(x[, start[, end]])
Purpose: Returns the index of the first occurrence of x in the list.
value → The element you want to find
start (optional) → Search from this index
end (optional) → Search up to this index
Returns: The index of the first match
Raises: ValueError if the element is not found
```

```python
l = [10, 20, 30, 20, 40,20]
l.index(20)
```

```python
l.index(20)  # Output: 1, because the first 20 is at index 1
l.index(20, 2)  # Output: 3, start searching from index 2
```

```python
#in 2d
#index() only matches entire elements, so it won’t find 5 inside [4,5,1]

l = [[4,5,1],[4,5]]
l.index([4,5,1])  # Output: 0
l.index([4,5])    # Output: 1
```

```python
suh=[1,2,3,4,5]   # l.copy()
aak=suh
print(id(suh))
print(id(aak))
suh[2]=20
print(suh)
print(aak)
```

```python
#in 1d
l=[1,2,3,4,5]   # l.copy()
l1=l.copy()
print(id(l))
print(id(l1))
l[2]=20
print(l1)
print(l)
```

```python
#in case of 2d
it create a new outer list. But the inner lists are still the same objects:
```

```python
l = [[1, 2], [3, 4]]
l1 = l.copy()
print(id(l1))
print(id(l))
print(l)
print(l1)
print(id(l1[1]))
print(id(l[1]))
```

```python
print(id(l1[0]))
l1[0][0] = 100
print(l1)
print(l)
print(id(l1[1]))
print(id(l[1]))
```

```python
A deep copy create a duplicates copy of all nested objects, 
so changes in the original list do not affect the copy.
```

```python
import copy
l = [[1, 2], [3, 4]]   # l.deepcopy()
l1= copy.deepcopy(l)
l1[0][0] = 500
print(l1) 
print(l)
print(id(l1[0]))
print(id(l[0]))
```

```python
#If you want to sort by string (alphabetical) order, convert all items to strings:
items = ["Apple", 10, "aanana", 2]
sorted_items = sorted(items, key=str)   # in ascending order
print(sorted_items) #"10" < "2" < "apple" < "banana" (lexicographically).
```

```python
#Sort numerically where possible, then alphabetically

If you want numbers first (sorted numerically) and then strings (sorted alphabetically):
```

```python
items = ["apple", 200, "banana", 20]

sorted_items = sorted(items, key=lambda x: (isinstance(x, str), str(x)))
print(sorted_items)
```

```python
items = [10, "2", "30"]
#print(sorted(items))  # ['10', '2', '30'] #That’s string order, not numeric order.
```

```python
#To sort by number value:
print(sorted(items, key=int,reverse=False))  # ['2', '10', '30']
```
